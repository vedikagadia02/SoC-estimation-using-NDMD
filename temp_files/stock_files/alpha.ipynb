```python
import requests
import pandas as pd
```

```python
url = 'https://www.alphavantage.co/query?function=LISTING_STATUS&apikey=O0TC2BS57NTIS0AF'
data = pd.read_csv(url)
stock_data = data[(data['assetType'] == 'Stock') & (data['exchange'] == 'NASDAQ')]
nasdaq_symbols = stock_data['symbol'].values
```

```python
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=IBM&interval=5min&apikey=O0TC2BS57NTIS0AF'
r = requests.get(url)
data = r.json()

df = pd.DataFrame(data['Time Series (Daily)']).T
df.columns = ['open', 'high', 'low', 'close', 'volume']
df = df.reset_index().rename(columns={'index': 'date'})
print(df)
df.to_csv('time_series_daily.csv')
```
